# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import symbols, integrate, exp, lambdify

x = symbols('x')
f = x * exp(-x**2)
g = x + 1

# Encontre os pontos de interseção das duas curvas
intersections = np.roots([1, -1, -1])
a = intersections[0]
b = 2

# Calcule a área entre as curvas
area1 = integrate(g - f, (x, a, b)).evalf()
area2 = integrate(g, (x, b, 2)).evalf()
area = area1 + area2

# Trace o gráfico das duas funções e preencha a área entre elas
x_vals = np.linspace(-3, 3, 100)
f_lambdified = lambdify(x, f)
g_lambdified = lambdify(x, g)
f_vals = f_lambdified(x_vals)
g_vals = g_lambdified(x_vals)

plt.plot(x_vals, f_vals)
plt.plot(x_vals, g_vals, color='purple')
plt.fill_between(x_vals, f_vals, g_vals, where=(g_vals > f_vals), interpolate=True, color='lightblue')
plt.fill_between([b, 2], [0, 0], [g_lambdified(b), g_lambdified(2)], interpolate=True)

plt.title(f'Área entre as curvas: {area:.2f}')
plt.show()
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definindo a função
x = sp.symbols('x')
f = x**2+1

# Intervalo de integração
a = 0
b = 2

# Calculando a área usando álgebra
area = sp.integrate(f, (x, a, b))

# Convertendo a função para uma função numpy
f_np = sp.lambdify(x, f)

# Plotando o gráfico da função
x_vals = np.linspace(a-2, b+2, 100)
y_vals = f_np(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.fill_between(x_vals, y_vals, where=((a-2 <= x_vals) & (x_vals <= b+2)), alpha=0.5)

# Marcando a área entre os limites de integração
plt.axvline(x=a, color='red', linestyle='--', label='Limite Inferior')
plt.axvline(x=b, color='green', linestyle='--', label='Limite Superior')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função')
plt.legend()
plt.show()

# Imprimindo o resultado da área
print("A área sobre a curva é:", area)
```

#Questão 4 (vale até 1 ponto)
A taxa de variação de juros nos primeiros meses do ano de 2023 é dada por $v = 3t^2 - e^t$. Onde $t$ é o tempo dado em dias. Encontre a função que determina o juro, $j$ a ser pago em um determinado dia deste período de forma algébrica em latex.



$$\int \:3t^2-e^tdt$$


$$\mathrm{Aplicar\:a\:regra\:da\:soma}:$$
$$\int \:3t^2dt-\int \:e^tdt$$



$$t^3-e^t$$

$$\mathrm{Adicionar\:uma\:constante\:à\:solução}:$$
$$t^3-e^t+C$$

```python
import sympy as sp

# Definir a variável simbólica t
t = sp.symbols('t')

# Definir a taxa de variação de juros v
v = 3*t**2 - sp.exp(t)

# Calcular a integral da taxa de variação v para obter a função do juro j
j = sp.integrate(v, t)

# Imprimir a função do juro
print('A função do juro é: j =', j)
```
